### FLIESS KAPPA CALCULATION FOR THE CV DATASET

```python
import pandas as pd
from statsmodels.stats.inter_rater import fleiss_kappa

# Extract the three choice columns
choice_column_1 = shivansh_cv["tag"]
choice_column_2 = jaya_cv["label"]

# orgnizing the data into a rating matrix
# Assuming categories are integers or labels and creating the rating matrix

categories = sorted(set(choice_column_1) | set(choice_column_2))
rating_matrix = []

for i in range(len(choice_column_1)):
  row = [0]* len(categories)
  row[categories.index(choice_column_1[i])] += 1
  row[categories.index(choice_column_2[i])] += 1
  rating_matrix.append(row)

# Convert to DataFrame for clarity
rating_df = pd.DataFrame(rating_matrix, columns=categories)

# Calculate Fleiss' Kappa
kappa = fleiss_kappa(rating_df.values, method='fleiss')
print(f"Fleiss' Kappa: {kappa}")
```

```python
import pandas as pd
from statsmodels.stats.inter_rater import fleiss_kappa

# Extract the three choice columns
choice_column_1 = shivansh_cv["tag"]
choice_column_2 = jaya_cv["label"]
choice_column_3 = outsider_cv["tag"]

# orgnizing the data into a rating matrix
# Assuming categories are integers or labels and creating the rating matrix

categories = sorted(set(choice_column_1) | set(choice_column_2) | set(choice_column_3))
rating_matrix = []

for i in range(len(choice_column_1)):
  row = [0]* len(categories)
  row[categories.index(choice_column_1[i])] += 1
  row[categories.index(choice_column_2[i])] += 1
  row[categories.index(choice_column_3[i])] += 1
  rating_matrix.append(row)

# Convert to DataFrame for clarity
rating_df = pd.DataFrame(rating_matrix, columns=categories)

# Calculate Fleiss' Kappa
kappa = fleiss_kappa(rating_df.values, method='fleiss')
print(f"Fleiss' Kappa: {kappa}")
```

```python
# Interpret the score
if kappa < 0:
    interpretation = "Poor agreement. Worse than chance."
elif 0 <= kappa <= 0.20:
    interpretation = "Slight agreement."
elif 0.21 <= kappa <= 0.40:
    interpretation = "Fair agreement."
elif 0.41 <= kappa <= 0.60:
    interpretation = "Moderate agreement."
elif 0.61 <= kappa <= 0.80:
    interpretation = "Substantial agreement."
elif 0.81 <= kappa <= 1.00:
    interpretation = "Almost perfect agreement."
else:
    interpretation = "Invalid score."

print(f"Fliess' Kappa Score: {kappa}")
print(f"Interpretation: {interpretation}")
```

```python
import json

# Function to sort a list of dictionaries by 'start'
def sort_ner(ner_entry):
    # If the entry is a string, convert it to a Python list
    if isinstance(ner_entry, str):
        ner_entry = json.loads(ner_entry)
    # Sort the list of dictionaries by the 'start' key
    return sorted(ner_entry, key=lambda x: x["start"])

# Apply the function to each element in the 'ner' column
shivansh_nlp['ner'] = shivansh_nlp['ner'].apply(sort_ner)
jaya_nlp['ner'] = jaya_nlp['ner'].apply(sort_ner)

# Display the updated DataFrame
print(shivansh_nlp['ner'][0])
print(jaya_nlp['ner'][2])
```

```python
import json
from sklearn.metrics import cohen_kappa_score

# Function to calculate Cohen's Kappa for two annotation datasets
def calculate_kappa(shivansh_nlp, jaya_nlp):
    shivansh_labels = []
    jaya_labels = []

    # Loop through each row of both datasets
    for shiv_ner, jaya_ner in zip(shivansh_nlp["ner"], jaya_nlp["ner"]):
        # Parse the `ner` JSON strings into lists of dictionaries


        # Extract spans (start, end, and label) into sets for each annotator
        shiv_set = {(entry["start"], entry["end"], entry["labels"][0]) for entry in shiv_ner}
        jaya_set = {(entry["start"], entry["end"], entry["labels"][0]) for entry in jaya_ner}

        # Create a union of spans to ensure we compare all annotations
        all_spans = shiv_set.union(jaya_set)

        # Encode presence/absence of each span as binary values
        for span in all_spans:
            shivansh_labels.append(1 if span in shiv_set else 0)
            jaya_labels.append(1 if span in jaya_set else 0)

    # Calculate Cohen's Kappa score
    return cohen_kappa_score(shivansh_labels, jaya_labels)

# Calculate Kappa
kappa_score = calculate_kappa(shivansh_nlp, jaya_nlp)
print(f"Cohen's Kappa Score: {kappa_score}")
```

```python
kappa = kappa_score
# Interpret the score
if kappa < 0:
    interpretation = "Poor agreement. Worse than chance."
elif 0 <= kappa <= 0.20:
    interpretation = "Slight agreement."
elif 0.21 <= kappa <= 0.40:
    interpretation = "Fair agreement."
elif 0.41 <= kappa <= 0.60:
    interpretation = "Moderate agreement."
elif 0.61 <= kappa <= 0.80:
    interpretation = "Substantial agreement."
elif 0.81 <= kappa <= 1.00:
    interpretation = "Almost perfect agreement."
else:
    interpretation = "Invalid score."

print(f"Cohen's Kappa Score: {kappa}")
print(f"Interpretation: {interpretation}")
```
